```python
import pandas as pd 
import numpy as np
import seaborn as  sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("compressed_data (1).csv")
```

```python
df.head(n=5)
```

```python
df.columns
```

```python
df.info()
```

# checking missing values 

```python
df.isnull().sum()
```

# changing format of last review  from object to date,time

```python
df["last review"]=pd.to_datetime(df["last review"] , errors="coerce") #coerce is to replace null with nat to handle error
```

```python
df.info()
```

# filling missing values with 0 and with minimum 

```python
df.fillna({"reviews per month" : 0 ,"last review" : df["last review"].min()}, inplace= True)
```

# removing missing values from NAME and host name column

```python
df.dropna(subset ={"NAME" ,"host name"},inplace=True)
```

```python
df.isnull().sum()
```

# removing useless columns from table

```python
df=df.drop(columns={"house_rules","license"},errors="ignore")
```

```python
df.columns
```

# Removing $ sign from columns which may cause errors further

```python
df["price"] = df["price"].replace("[\$,]","", regex=True).astype(float)
df["service fee"]=df["service fee"].replace("[$,]","", regex=True).astype(float)
```

```python
df.head()
```

# remove duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# Descriptive statistics

```python
df.describe()
```

# Data visualization

```python
#Distribution of listing prices?
plt.figure(figsize=(6,10))
sns.histplot(df["price"],bins=50,kde=True,color="blue")
plt.title("Distribution of listing price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

```python
"""conclusion:-The distribution of listing prices appears fairly uniform across the range, with no strong skewness or extreme peaks.
1)Prices are spread evenly.
2)No major outliers or sharp spikes.
3)Slight fluctuations, but overall stable distribution."""
```

# How are different room types distributed?

```python
df["room type"]
```

```python
plt.figure(figsize=(8,5))
sns.countplot( x="room type",data=df,color="red")
plt.title("Room type distribution")
plt.xlabel("Room types")
plt.ylabel("Frequency")
plt.show()
```

# How are listings distributed across different neighborhoods?

```python
plt.figure(figsize=(12,8))
sns.countplot( y="neighbourhood group",data=df,color="lightgreen",order=df["neighbourhood group"].value_counts().index)
plt.title("number of listing by Neighbourhood Group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood Group")
plt.show()
```

# What is the relationship between price and room type ?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room type",y="price",hue="room type",data=df,palette="Set1")
plt.title("Price vs. Room type")
plt.xlabel("Room type")
plt.ylabel("Price ($)")
plt.legend(title="Room title")
plt.show()
```

# How has the number of reviews change over with time ?

```python
df["last review"] = pd.to_datetime(df["last review"])
reviews_over_time = df.groupby(df["last review"].dt.to_period("M")).size()


plt.figure(figsize=(12,6))
reviews_over_time.plot(kind="line",color="red")
plt.title("Number of Reviews Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.show()
```
